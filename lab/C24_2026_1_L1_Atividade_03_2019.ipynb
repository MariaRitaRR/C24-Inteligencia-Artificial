## <font color='green'>**2.5 Identificação de Problemas nos Dados**</font>

### <font color='cyan'> Q16: Inspecione o conjuntos de dados para verificar se há:</font>
- #### <font color='cyan'> Dados Ausentes;</font>
- #### <font color='cyan'> Valores Atípicos (Outliers);</font>
- #### <font color='cyan'> Valores Duplicados;</font>
### <font color='cyan'> Reporte sua(s) descoberta(s) a seguir.</font>

```python
# DADOS AUSENTES
ausentes = df.isnull().sum()
print(ausentes[ausentes > 0])
print(f"\nTotal de valores ausentes: {df.isnull().sum().sum()}")

# VALORES DUPLICADOS
duplicados = df.duplicated().sum()
print(f"Linhas duplicadas: {duplicados}")
if duplicados > 0:
    print(f"Percentual: {(duplicados/len(df))*100:.2f}%")

# OUTLIERS (VALORES ATÍPICOS)

print("\nEstatísticas descritivas (observe min e max muito distantes da média):")
print(df.describe())


print("\nPossíveis outliers por coluna numérica:")
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = df[(df[col] < limite_inferior) | (df[col] > limite_superior)]

    if len(outliers) > 0:
        print(f"\n{col}:")
        print(f"  - Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
        print(f"  - Limites: [{limite_inferior:.2f}, {limite_superior:.2f}]")
        print(f"  - Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
```

O sinal positivo mostra que, em geral, quando uma das variáveis aumenta, a outra tende a aumentar também. Como o valor está em torno de 0,5, a relação entre elas não é fraca, mas também não é muito forte, caracterizando uma associação moderada entre as duas características.
